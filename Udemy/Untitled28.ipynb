```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans

data=pd.read_csv(r'C:\\Users\\LasimaSN\\Downloads\\The Data Science Course 2020 - All Resources\\Part_5_Advanced_Statistical_Methods_(Machine_Learning)\S38_L257\Countries-exercise.csv')
data
```

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

```python
x=data.iloc[:,1:]
x
```

```python
km=KMeans(2)
km.fit(x)
identify_clusters=km.fit_predict(x)
```

```python
new_data=data.copy()
new_data['Clusters']=identify_clusters
new_data
```

```python
plt.scatter(new_data['Longitude'],new_data['Latitude'],c=new_data['Clusters'],)
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
