```python
import numpy as np
import pandas as pd
import statsmodels.api as st
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
data=pd.read_csv(r'C:\Users\LasimaSN\Downloads\The Data Science Course 2020 - All Resources\Part_5_Advanced_Statistical_Methods_(Machine_Learning)\S33_L204\1.03. Dummies.csv')
```

```python
data
```

```python
rdata= data.copy()
rdata['Attendance'] = rdata['Attendance'].map({'Yes':1,'No':0})
```

```python
rdata
```

```python
rdata.describe()
```

```python
y=rdata['GPA']
x1=rdata[['SAT','Attendance']]
```

```python
x=st.add_constant(x1)
results=st.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(rdata['SAT'],y)
y_no= 0.6439 + 0.0014*rdata['SAT']
y_yes= 0.8665+ 0.0014*rdata['SAT']
fig = plt.plot(rdata['SAT'],y_no,lw=2,c='Green')
fig = plt.plot(rdata['SAT'],y_yes,lw=2,c='Red')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```
